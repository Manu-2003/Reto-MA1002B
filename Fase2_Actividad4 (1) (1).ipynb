```python
import math  

def func_x( x, y): 
    return (x*(1-(x/500))-0.02*x*y) 
def func_y( x, y ): 
    return (-y+0.01*x*y) 

def euler( t0, y0, x0, h, tf ): 
    valuesx = []
    valuesy = []
    valuest = []
    while t0 < tf: 
        y0 = y0 + h * func_y(x0, y0) 
        x0 = x0 + h * func_x(x0, y0) 
        t0 = t0 + h 
        valuesx.append(x0)
        valuesy.append(y0)
        valuest.append(t0)
    return (valuesx,valuesy,valuest)    
```

```python
x,y,t = euler(0, 20, 20, 0.01, 30) 
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
plt.plot(x,y, 'g',label='solución numérica')
ax.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
plt.plot(t,x, 'b', label = 'Presas')

plt.plot(t,y, 'r', label = "Depredadores")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import integrate
import ipywidgets as ipw

alpha = 1.
beta = 0.02
delta = 1
gamma = 0.01
x_0 = 20
y_0 = 20

def derivative(X, t, alpha, beta, delta, gamma):
    x1, y1 = X
    dotx =x1*(1-(x1/500))-0.02*x1*y1
    doty = -y1+0.01*x1*y1 
    return np.array([dotx, doty])

Nt = 3000
tmax = 30
t1 = np.linspace(0,tmax, Nt)
X0 = [x_0, y_0]
res = integrate.odeint(derivative, X0, t1, args = (alpha, beta, delta, gamma))
xt, yt = res.T

plt.figure()
plt.grid()
plt.title("odeint method")
plt.plot(t1, xt, 'm-', label = 'Presas')
plt.plot(t1, yt, 'r-', label = "Depredadores")
plt.xlabel('Tiempo')
plt.ylabel('Población')
plt.legend()

plt.show()
len(t1)
```
